```python
# write a program to chech if a list cantains a palindrome of elements.(hint:use copy() method.)
list1=[1,2,1]

copy_list1=list1.copy()
copy_list1.reverse()

if(copy_list1==list1):
    print("palindrome")
else:
    print("not palindrome")
```

```python
list2=[1,2,3]

copy_list2=list2.copy()
copy_list2.reverse()

if(copy_list2==list2):
    print("palindrome")
else:
    print("not palindrome")
```

```python
num=[1,2,3,2,1]

copy_num=num.copy()
copy_num.reverse()

if(copy_num==num):
    print("palindrome")
else:
    print("not palindrome")
```
